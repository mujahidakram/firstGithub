```python
#Importing Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#importing data set
palette=sns.color_palette('rocket_r')
df=pd.read_csv(r'C:\Users\Mujahid\Desktop\python_with_aammar\Chilla_data2_for_plots.csv')
p=sns.relplot(x="Age (years)-Float/Int", y="Your Weight in kg? (float)", hue="How many hours you code a day? (int) e.g: 5,4,3", 
              col="Gender", 
              kind="line", palette=palette, data=df )
plt.show()
```

```python
#Importing Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=df,
    x="Age (years)-Float/Int", y="Your Weight in kg? (float)",
     size="Marital Status?", col="Gender", 
    kind="line", size_order=["Yes", "No"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```
