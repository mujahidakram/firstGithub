# How to add title to your graph

```python
#Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


#load Data Set

iris=sns.load_dataset("iris")

sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("IRIS Sepal Length and Width Graph")
plt.show()
```

## How to increase or decrease area of plot

```python
#Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


#load Data Set

iris=sns.load_dataset("iris")

sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
#area under consideration use xlim and ylim
plt.xlim(3)
plt.ylim(1)

plt.title("IRIS Sepal Length and Width Graph")
plt.show()
```

## How to add styles to your plot
### types of styles in seaborn
- whitegrid
- darkgrid
- white
- dark
- ticks


```python
#Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


#load Data Set
sns.set_style("whitegrid")

iris=sns.load_dataset("iris")

sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("IRIS Sepal Length and Width Graph")
plt.show()
```

```python
#Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


#load Data Set
sns.set_style("whitegrid")

iris=sns.load_dataset("iris")

sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("IRIS Sepal Length and Width Graph")
plt.show()
```

### Bar Plots in Python

```python
#Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


#load Data Set
sns.set_style("whitegrid")

iris=sns.load_dataset("iris")

sns.barplot(x="species", y="sepal_width", data=iris)
plt.title("IRIS Sepal Length and Width Graph")
plt.show()
```

```python
#Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


#load Data Set
sns.set_style("whitegrid")

iris=sns.load_dataset("iris")

sns.barplot(x="species", y="sepal_width", data=iris, color="red")
plt.title("IRIS Sepal Length and Width Graph")
plt.show() 
```

```python
#Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


#load Data Set
sns.set_style("whitegrid")

titanic=sns.load_dataset("titanic")

sns.barplot(x="sex", y="alone", hue="who" , data=titanic )
plt.title("TITANIC Barplot")
plt.show() 
```

```python
#Import Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


#load Data Set
sns.set_style("whitegrid")

titanic=sns.load_dataset("titanic")

sns.barplot(x="sex", y="alone", hue="who" , data=titanic, order=["female", "male"], color="teal")
plt.title("TITANIC Barplot")
plt.show() 
```
